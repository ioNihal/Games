```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, binom, poisson, uniform, logistic
def plot_distribution(distribution, title):
   plt.hist(distribution, bins=30, density=True)
   plt.title(title)
   plt.show()
plot_distribution(np.random.normal(0,0.1, 1000), 'Normal Distribution')
plot_distribution(np.random.binomial(10, 0.5, 1000), 'Binomial Distribution')
plot_distribution(np.random.poisson(5, 1000), 'Poisson Distribution')
plot_distribution(np.random.uniform(0, 1, 1000), 'Uniform Distribution')
plot_distribution(np.random.logistic(0, 1, 1000), 'Logistic Distribution')
```
